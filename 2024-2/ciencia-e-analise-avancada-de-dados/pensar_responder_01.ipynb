```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import binom
```

```python
# Parâmetros da distribuição binomial
n = 1000
p = 0.5

x = np.arange(0, n + 1)

binomial = binom.pmf(k=x, n=n, p=p)

plt.bar(x, binomial)

plt.xlabel("x", fontsize=12)
plt.ylabel("Probabilidade", fontsize=12)
plt.xlim([-1, n + 1])

plt.title(f"Distribuição Binomial, n={n} p={p}", fontsize=15)
plt.show()
```

```python
# Alterando os parâmetros da distribuição binomial
n = 100

binomial = binom.pmf(k=x, n=n, p=p)

plt.bar(x, binomial)
plt.show()
```
